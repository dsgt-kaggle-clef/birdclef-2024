```python
from pathlib import Path

soundscapes = sorted(
    Path("/mnt/data/raw/birdclef-2024/unlabeled_soundscapes").glob("*.ogg")
)
print(soundscapes[:3])
```

```python
from birdnetlib import Recording, RecordingBuffer
from birdnetlib.analyzer import Analyzer

analyzer = Analyzer()
recording = Recording(analyzer, soundscapes[0], overlap=2)
recording.extract_embeddings()
print(recording.embeddings)
```

```python
[(x["start_time"], x["end_time"]) for x in recording.embeddings][:10]
```

```python
# let's read using torch audio instead and see how that works out
import torchaudio

audio, _ = torchaudio.load(soundscapes[0])
audio.shape
```

```python
recording = RecordingBuffer(
    analyzer, audio.squeeze().numpy()[: 5 * 48000], 48000, overlap=1
)
recording.extract_embeddings()
[(x["start_time"], x["end_time"]) for x in recording.embeddings]
```
